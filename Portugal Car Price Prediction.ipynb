```python
def remove_outliers(df, feature):
    #InterQuantile Range: 75% - 25% values in a dataset

    # Steps:
    #  1. Arrange the data in increasing order
    dataset=df
    dataset.sort_values(by=[feature], inplace = True)

    #  2. Calculate first(q1) and third(q3)
    quantile1, quantile3 = np.percentile(dataset[feature],[25,75])
    print ('Quantile 1:', quantile1)
    print ('Quantile 1:', quantile3)

    #  3. Find interquartile range (q3-q1)
    iqr_value = quantile3 - quantile1
    print ('Interquartile range:', iqr_value)

    #  4. Find lower bound q1*1.5
    lower_bound_val = quantile1 - (1.5 * iqr_value)

    #  5. Find upper bound q3*1.5
    upper_bound_val = quantile3 + (1.5 * iqr_value)
    print('Lower bound:', lower_bound_val)
    print('Upper bound:', upper_bound_val)
    
    init_len = df.shape[0]
    df.drop(df[(df[feature] >= upper_bound_val) | (df[feature] <= lower_bound_val)].index, inplace=True)
    final_len = df.shape[0]
    print(init_len - final_len, "entries were dropped")

    return df
```

```python
def detect_outliers(data):
    outliers =[]
    threshold = 3 #for 99.73% acurancy
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        #print (str(i) + ', Z score: ' + str(z_score) )
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return len(outliers), outliers
```

```python
def run_model_LR(X, y, lr_fit_intercept):
    """Linear Regression"""
    model = LinearRegression(fit_intercept=lr_fit_intercept).fit(X,y)
    return model

def run_model_Lasso(X, y, lasso_alpha):
    """Linear Lasso Model"""
    model = linear_model.Lasso(alpha=lasso_alpha).fit(X, y)
    return model

def run_model_Ridge(X, y, ridge_alpha):
    """Linear Ridge Model"""
    model = linear_model.Ridge(alpha=ridge_alpha).fit(X, y)
    return model

def run_model_DT(X, y, max_depth=None, min_samples_split=2, min_samples_leaf=1):
    """Decision Tree Regression"""
    model = DecisionTreeRegressor(max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf).fit(X, y)
    return model

def run_model_RF(X, y, n_estimators=100, max_depth=None, min_samples_split=2, min_samples_leaf=1):
    """Random Forest Regression"""
    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf).fit(X, y)
    return model

def run_model_XGB(X, y, max_depth=3, learning_rate=0.1, n_estimators=100):
    """XGBoost Regression"""
    model = xgb.XGBRegressor(max_depth=max_depth, learning_rate=learning_rate, n_estimators=n_estimators).fit(X, y)
    return model
```

```python
df_model_scores = pd.DataFrame(columns=['Model', 'MAE', 'MSE', 'R2'])

def cal_model_scores(model, y_pred, y_test):
    "Calculate model's MAE, MSE and R2"
    mae = round(mean_absolute_error(y_test, y_pred), 3)
    mse = round(mean_squared_error(y_test, y_pred), 3)
    r2 = round(r2_score(y_test, y_pred), 3)
    scores = [model, mae, mse, r2]
    
    return scores
```

```python
def add_scores_to_df(df_model_scores, scores):
    model_scores = pd.DataFrame(scores, columns=['Model', 'MAE', 'MSE', 'R2'])
    model_scores = model_scores.groupby('Model').mean().round(2).reset_index()
    df_model_scores = pd.concat([df_model_scores, model_scores], ignore_index=True).sort_values(by=['MAE'])
    
    return df_model_scores
```

```python
# Functions to return the average score between training and test for the various models 
# The function uses the specified cross-validation method to split the data into training and test sets multiple times

def avg_score_LR(method, X, y, df_model_scores, lr_fit_intercept):
    """Calculate the average of MAE, MSE and R2 from the Linear Regression model.
       Use the best parameters found in GridSearchCV"""
    scores = []
    for train_index, test_index in method.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model = run_model_LR(X_train, y_train, lr_fit_intercept)
        lr_y_pred = model.predict(X_test)        
        score = cal_model_scores('LR', lr_y_pred, y_test)
        scores.append(score)

    df_model_scores = add_scores_to_df(df_model_scores, scores)
    
    return model, df_model_scores

def avg_score_Lasso(method, X, y, df_model_scores, lasso_alpha):
    """Calculate the average of MAE, MSE and R2 from the Lasso model.
       Use the best parameters found in GridSearchCV"""
    scores = []
    for train_index, test_index in method.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model = run_model_Lasso(X_train, y_train, lasso_alpha)
        lr_y_pred = model.predict(X_test)        
        score = cal_model_scores('Lasso', lr_y_pred, y_test)
        scores.append(score)
        
    model_scores = pd.DataFrame(scores, columns=['Model', 'MAE', 'MSE', 'R2'])
    model_scores = model_scores.groupby('Model').mean().round(2).reset_index()
    df_model_scores = pd.concat([df_model_scores, model_scores], ignore_index=True).sort_values(by=['MAE'])
    
    return model, df_model_scores

def avg_score_Ridge(method, X, y, df_model_scores, ridge_alpha):
    """Calculate the average of MAE, MSE and R2 from the Ridge model.
       Use the best parameters found in GridSearchCV"""
    scores = []
    for train_index, test_index in method.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model = run_model_Ridge(X_train, y_train, ridge_alpha)
        lr_y_pred = model.predict(X_test)        
        score = cal_model_scores('Ridge', lr_y_pred, y_test)
        scores.append(score)
        
    model_scores = pd.DataFrame(scores, columns=['Model', 'MAE', 'MSE', 'R2'])
    model_scores = model_scores.groupby('Model').mean().round(2).reset_index()
    df_model_scores = pd.concat([df_model_scores, model_scores], ignore_index=True).sort_values(by=['MAE'])
    
    return model, df_model_scores

def avg_score_DT(method, X, y, df_model_scores, max_depth=None, min_samples_split=2, min_samples_leaf=1):
    """Calculate the average of MAE, MSE and R2 from the Decision Tree Regression model.
       Use the best parameters found in GridSearchCV"""
    scores = []
    for train_index, test_index in method.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model = run_model_DT(X_train, y_train, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)
        dt_y_pred = model.predict(X_test)
        score = cal_model_scores('DT', dt_y_pred, y_test)
        scores.append(score)

    df_model_scores = add_scores_to_df(df_model_scores, scores)

    return model, df_model_scores

def avg_score_RF(method, X, y, df_model_scores, n_estimators=100, max_depth=None, min_samples_split=2, min_samples_leaf=1):
    """Calculate the average of MAE, MSE and R2 from the Random Forest Regression model.
       Use the best parameters found in GridSearchCV"""
    scores = []
    for train_index, test_index in method.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model = run_model_RF(X_train, y_train, n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)
        rf_y_pred = model.predict(X_test)
        score = cal_model_scores('RF', rf_y_pred, y_test)
        scores.append(score)

    df_model_scores = add_scores_to_df(df_model_scores, scores)

    return model, df_model_scores

def avg_score_XGB(method, X, y, df_model_scores, max_depth=3, learning_rate=0.1, n_estimators=100):
    """Calculate the average of MAE, MSE, and R2 from the XGBoost Regression model.
       Use the best parameters found in GridSearchCV"""
    scores = []
    for train_index, test_index in method.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model = run_model_XGB(X_train, y_train, max_depth=max_depth, learning_rate=learning_rate, n_estimators=n_estimators)
        xgb_y_pred = model.predict(X_test)
        score = cal_model_scores('XGB', xgb_y_pred, y_test)
        scores.append(score)

    df_model_scores = add_scores_to_df(df_model_scores, scores)

    return model, df_model_scores
```
