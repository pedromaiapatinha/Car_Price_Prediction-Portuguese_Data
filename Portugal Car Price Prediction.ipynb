### Outliers
- High Valued Cars
- Classic Cars

```python
def detect_outliers(data):
    outliers =[]
    threshold = 3 #for 99.73% acurancy
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        #print (str(i) + ', Z score: ' + str(z_score) )
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return len(outliers), outliers
```

```python
for column in df.columns:
    if df[column].dtypes not in ('object', 'datetime64[ns]'):
        print(column, detect_outliers(df[column]))
```

## Models

```python
df_scores = pd.DataFrame(columns=['Model', 'MAE', 'MSE', 'R2'])

def add_scores(model, y_pred, y_test):
    mae = round(mean_absolute_error(y_test, y_pred), 3)
    mse = round(mean_squared_error(y_test, y_pred), 3)
    r2 = round(r2_score(y_test, y_pred), 3)
    scores = [model, mae, mse, r2]
    df_scores.loc[df_scores.shape[0]] = scores
    df_scores.sort_values(by=['MAE'], inplace=True)
    
    return df_scores
```

```python
parameter_space = {
    'hidden_layer_sizes': [(50,50,50), (100,)],
    'activation': ['logistic', 'tanh', 'relu'],
    'solver': ['lbfgs', 'sgd', 'adam'],
    'learning_rate_init': list(np.linspace(0.00001,0.1,5)),
    'learning_rate': ['constant','adaptive']
}
```

```python
model = MLPRegressor(max_iter=200)
clf = GridSearchCV(model, parameter_space)
clf.fit(X_train, y_train)
```

```python
# Best parameter set
print('------------------------------------------------------------------------------------------------------------------------')
print('Best parameters found:\n', clf.best_params_)
print('------------------------------------------------------------------------------------------------------------------------')

# All results
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std , params))
```
